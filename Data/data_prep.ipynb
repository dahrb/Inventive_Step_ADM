```python
from pdf2image import convert_from_path
from PIL import Image
import pytesseract
import re

#use tesseract app
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

BASE_DIR = "../Data/TEST"
DPI = 300
TESS_CONFIG = "--psm 3"

def light_clean(text):
    # minimal cleaning: unify line endings, collapse internal whitespace, keep paragraph breaks
    text = text.replace("\r\n", "\n").replace("\r", "\n")
    lines = [re.sub(r"\s+", " ", ln).rstrip() for ln in text.split("\n")]
    # remove leading/trailing blank lines
    while lines and lines[0] == "":
        lines.pop(0)
    while lines and lines[-1] == "":
        lines.pop()
    return "\n".join(lines)

def process_pdf_to_text(pdf_path):
    pages = convert_from_path(pdf_path, dpi=DPI)
    page_texts = []
    for i, page in enumerate(pages, start=1):
        raw = pytesseract.image_to_string(page, config=TESS_CONFIG)
        cleaned = light_clean(raw)
        page_texts.append(f"--- PAGE {i} ---\n{cleaned}")
    full_text = "\n\n".join(page_texts)
    full_text = re.sub(r'\n{3,}', '\n\n', full_text).strip()  # collapse excessive blank lines
    return full_text

# iterate each case folder, prefer claims.pdf then application.pdf, write to claims.txt (outside raw/)
for case_name in sorted(os.listdir(BASE_DIR), key=lambda s: s.lower()):
    case_dir = os.path.join(BASE_DIR, case_name)
    if not os.path.isdir(case_dir):
        continue
    raw_dir = os.path.join(case_dir, "raw")
    if not os.path.isdir(raw_dir):
        continue
    
    #print(case_name)

    for candidate in ("claim.pdf", "application.pdf"):
        pdf_path = os.path.join(raw_dir, candidate)
        #print(pdf_path)
        if os.path.isfile(pdf_path):
            text = process_pdf_to_text(pdf_path)
            out_path = os.path.join(case_dir, "claims.txt")
            try:
                with open(out_path, "w", encoding="utf-8") as fw:
                    fw.write(text)
                print("Wrote:", out_path)
            except Exception as e:
                print("Failed to write", out_path, ":", e)
            break
    
    else:
        print("No claims/application PDF found for", case_name)

```
