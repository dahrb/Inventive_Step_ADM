```python
import pandas as pd
import os
```

```python
#import previously processed data
inv_step = pd.read_pickle('/users/sgdbareh/scratch/ADM_JURIX/Data/Inv_Step_Test.pkl')
```

```python
#create the folder and file structure for collecting the data
base_dir = '../Data/TEST'

for idx, row in inv_step.iterrows():
    reference = str(row['Reference'])
    summary_facts = str(row['Summary Facts']) if 'Summary Facts' in row else ''
    folder_path = os.path.join(base_dir, reference)
    raw_folder = os.path.join(folder_path, 'raw')
    
    #create main and raw folders
    os.makedirs(raw_folder, exist_ok=True)
    
    #create and write to appeal.txt using the summary of facts section
    with open(os.path.join(folder_path, 'appeal.txt'), 'w') as f:
        f.write(summary_facts)
    
    #create empty claims.txt and CPA.txt
    open(os.path.join(folder_path, 'claims.txt'), 'w').close()
    open(os.path.join(folder_path, 'CPA.txt'), 'w').close()
```

```python
# Create a new DataFrame with the desired columns
excel_df = pd.DataFrame({
    "Reference": inv_step["Reference"],
    "CPA": "",
    "Claims": "",
    "Notes": ""
})

# Save to Excel
excel_df.to_excel("/users/sgdbareh/scratch/ADM_JURIX/Data/case_index.xlsx", index=False)
```
